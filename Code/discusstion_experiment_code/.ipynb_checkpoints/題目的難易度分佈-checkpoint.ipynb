本程式分析並呈現學測題目的正確答案詞頻難度以及混淆項的平均詞頻難度

詞頻難度來源: 高中英文參考詞彙表所訂定的1至6級難度

```python
from utils import init_data, lemmatization
import spacy

data, ref_word = init_data()
_ref_word = list(ref_word['單字'])
model = spacy.load('en_core_web_sm')

notInRef = []
hardness = []
distractor_hardness = []
d_in_ref = []
hardness_difference_between_ans_and_dis = []

for qa_pair in data:
    answer_word = lemmatization(qa_pair["correct_option_en"], language_model=model)
    dis = qa_pair['distractors']
    if answer_word in _ref_word:
        index = _ref_word.index(answer_word)
        ans_hardness = list(ref_word['難度'])[index]
        hardness.append(ans_hardness)
        d_hard = []
        for d in dis:
            d = lemmatization(d, language_model=model)
            if d in _ref_word:
                d_in_ref.append(d)
                index = _ref_word.index(d)
                d_hard.append(list(ref_word['難度'])[index])
        if len(d_hard) != 0:
            for dd in d_hard:
                hardness_difference_between_ans_and_dis.append(abs(dd-ans_hardness))
    else:
        notInRef.append(answer_word)

print(f"Answers that does not appear in reference: {notInRef}")
print(f"{len(notInRef)}/{len(data)}")
print(len(distractor_hardness))
```

```python
# 圖: target word的難易度分布

from utils import draw_plot
import numpy as np

print(f"平均: {sum(hardness)/len(hardness)}")
print(f"標準差: {np.std(hardness)}")
# Define the data to plot
plot_data = {'1': hardness.count(1), '2': hardness.count(2), '3': hardness.count(3), '4': hardness.count(4),'5':hardness.count(5),'6':hardness.count(6)}
draw_plot(plot_data, title="Difficulty distribution of correct answer", x_label = "difficulty", y_label="Quantity")
```

```python
# 圖: 同一題中正解與混淆項難度的差距
import numpy as np

print(hardness_difference_between_ans_and_dis)
data = {x:hardness_difference_between_ans_and_dis.count(x) for x in range(0, 6)}
print(data)

draw_plot(data, title="Difficulty difference between distractors and correct answer", x_label = "difficulty", y_label="Quantity")
```

# 結論
- 總體正確答案與混淆項的難易度分布大致接近，且由圖三可知道同一題的四個選項其難易度接近
- 得到了出題的難度分佈，可用於後續出題的難度分佈參考
